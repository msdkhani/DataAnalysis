# Importing Library

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

First of all want to real our dataset

```python
smokers=pd.read_csv("smokers.csv")
```

Here we want to get a description of our data base. To see what is happening

```python
smokers.describe()
```

As we can see above the "Sl no" does not help us to find any pattern or relation between other features so we can easily delete it

```python
smokers = smokers.drop(['Sl no'], axis=1)
smokers.head()
```

## Data preprocessing

#### Data pre-processing can be done in several stage

First of all we check to see if we have any duplicated item in our dataset

```python
smokers.duplicated().shape
```

as we can see above we do not have any duplicated value.

So next step is to check for any null or missing value in our data

```python
smokers.isnull().sum()
```

# Getting more information about our data


```python
smokers.dtypes
```

```python
smokers.nunique()
```

```python
sns.pairplot(smokers)
```

```python
plt.figure(figsize = (20,7))
smokers['Bone loss (0 and 1)'].value_counts().plot.bar(color = 'red')
plt.title('Checking How many patients have Bone loss')
plt.xlabel('No Bone loss v/s have bone loss')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (20,7))
smokers['Group'].value_counts().plot.bar(color = 'lime')
plt.title('Number of patients in different group')
plt.xlabel('Group')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (20,7))
smokers['smoking (Y/N)'].value_counts().plot.bar(color = 'blue')
plt.title('Checking How many patients are smoking')
plt.xlabel('Smoking v/s no smoking')
plt.ylabel('Count')
plt.show()
```

## Correlation
